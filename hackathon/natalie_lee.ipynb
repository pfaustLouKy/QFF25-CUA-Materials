```python
import numpy as np
from qiskit.quantum_info import SparsePauliOp

# 1. Define the Pauli string for the operator Z on qubit 2, Y on qubit 1, and I on qubit 0.
# Qiskit uses little-endian: index 2 | index 1 | index 0
#                           (Z)     |   (Y)   |   (I)
pauli_string = "ZYI"


# 2. Create the 3-qubit Pauli operator
# The second argument is the coefficient, which is 1.0 for a simple Pauli operator.
operator = SparsePauliOp.from_list([(pauli_string, 1.0)])

# 3. Print the operator
print("--- Pauli Operator Representation ---")
print(operator)
print("\n" + "-"*45 + "\n")

# 4. Print its corresponding matrix representation
# The .to_matrix() method returns the dense matrix representation (a 2^n x 2^n numpy array)
matrix_representation = operator.to_matrix()
print("--- Matrix Representation (8x8 numpy array) ---")
print(matrix_representation)

# Optional: Print the shape to confirm it's an 8x8 matrix (2^3 x 2^3)
print(f"\nMatrix shape: {matrix_representation.shape}")
```

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

def generate_pi_over_4_state():
    """
    Creates a 1-qubit circuit, puts the qubit in |1>, applies a pi/4 phase shift, 
    and outputs the resulting statevector in Dirac notation.
    """
    # Create a quantum circuit with 1 qubit
    qc = QuantumCircuit(1, name="PiOver4Phase")

    # 1. Put the qubit in the state |1>
    # The circuit starts in |0>. Applying an X gate flips it to |1>.
    qc.x(0)
    
    # 2. Add a single gate that applies a pi/4 phase shift
    # The Phase gate P(phi) applies a phase e^(i*phi) to the |1> component.
    # P(phi) = [[1, 0], [0, e^(i*phi)]]
    pi_over_4 = np.pi / 4
    qc.p(pi_over_4, 0)

    # Print the circuit structure
    print("--- Quantum Circuit Diagram ---")
    print(qc.draw(output='text', fold=-1))
    print("-" * 35)

    # 3. Calculate the final statevector
    initial_state = Statevector.from_int(0, 2**1) # Start from |0>
    final_state = initial_state.evolve(qc)

    # 4. Output the Dirac notation representation
    # State |0> is 0, State |1> is e^(i*pi/4)
    # The X gate changes |0> to |1>, so the state before P gate is |1>.
    # The P(phi) gate acts on |1> as P(phi)|1> = e^(i*phi)|1>
    
    # We use the final_state object's to_dict() method for a clean output.
    state_dict = final_state.to_dict()
    
    # Manually construct the Dirac notation for clarity
    # Since the state started as |0> -> X -> |1> -> P(pi/4) -> e^(i*pi/4) |1>
    
    amplitude_1 = final_state.data[1] # The coefficient for |1>
    
    # Format the complex number e^(i*pi/4) = cos(pi/4) + i*sin(pi/4) = (1/sqrt(2)) + i(1/sqrt(2))
    # Qiskit stores the state as [alpha, beta]. Here alpha=0 and beta=e^(i*pi/4).
    formatted_amplitude = f"{np.cos(pi_over_4):.4f} + {np.sin(pi_over_4):.4f}i"

    print("--- Final Statevector (Dirac Notation) ---")
    print(f"|ψ⟩ = {formatted_amplitude}|1⟩")
    print("-" * 35)
    
    # Optional: Print the full Statevector object data
    print("\nStatevector (Qiskit Array Format):")
    print(final_state.data)

# Run the function
generate_pi_over_4_state()
```

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
# The bloch sphere drawer is imported from qiskit.visualization
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

def solve_ry_probability_problem():
    """
    Creates a 1-qubit circuit with an RY gate to achieve the specified 
    measurement probabilities and displays the Bloch sphere.
    """
    # --- 1. Define Target Probabilities and Calculate Rotation Angle ---
    
    P_1_target = 0.854  # 85.4% probability of measuring |1>
    P_0_target = 0.146  # 14.6% probability of measuring |0>
    
    # Amplitude of |1> state is sin(theta/2)
    sin_theta_half = np.sqrt(P_1_target)
    
    # Calculate theta/2 angle in radians
    theta_half = np.arcsin(sin_theta_half)
    
    # Calculate the full rotation angle theta
    theta = 2 * theta_half

    print("--- Calculated Angle ---")
    print(f"Angle required (θ): {theta:.4f} radians ({np.degrees(theta):.2f} degrees)")
    print("-" * 35)


    # --- 2. Create and Evolve Quantum Circuit ---
    
    qc = QuantumCircuit(1, name="ProbState")
    
    # Apply the calculated RY rotation to qubit 0 (starts in |0>)
    qc.ry(theta, 0)

    # Print the circuit structure
    print("--- Quantum Circuit Diagram ---")
    print(qc.draw(output='text', fold=-1))
    print("-" * 35)

    # Calculate the resulting Statevector
    initial_state = Statevector.from_int(0, 2**1) 
    final_state = initial_state.evolve(qc)

    # Print the final statevector in array format
    print("--- Final Statevector (Amplitudes) ---")
    # Statevector is [alpha, beta] where alpha = cos(theta/2) and beta = sin(theta/2)
    print(final_state.data)
    print("-" * 35)


    # --- 3. Prints the Probabilities ---
    
    # Calculate the probabilities from the statevector
    # P(|0>) = |alpha|^2, P(|1>) = |beta|^2
    probabilities = final_state.probabilities()
    
    P0_measured = probabilities[0] * 100
    P1_measured = probabilities[1] * 100

    print("--- Measurement Probabilities ---")
    print(f"P(|0⟩): {P0_measured:.1f}% (Target: {P_0_target*100:.1f}%)")
    print(f"P(|1⟩): {P1_measured:.1f}% (Target: {P_1_target*100:.1f}%)")
    print("-" * 35)


    # --- 4. Displays a Bloch sphere representation ---
    
    # Use plot_bloch_multivector (works for single qubit statevector too)
    bloch_fig = plot_bloch_multivector(final_state)
    
    # Save or display the figure (required for some environments)
    # Note: In a Jupyter/Colab environment, this plot will be displayed automatically.
    # bloch_fig.show() # Use this if necessary to force display

    return bloch_fig

# Run the function
bloch_sphere_plot = solve_ry_probability_problem()

# Display the Bloch Sphere plot outside the main function body
# 
plt.show(bloch_sphere_plot)
```

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
# The bloch sphere drawer is imported from qiskit.visualization
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

p1 = 0.854                   # desired probability of |1>
amp = np.sqrt(p1)            # sin(theta/2)

# solve for theta/2
theta_over_2 = np.arcsin(amp)

# multiply by 2 to get theta
theta = 2 * theta_over_2

print("theta (radians):", theta)
print("theta (degrees):", np.degrees(theta))

qc = QuantumCircuit(1)
qc.h(0)
sv = Statevector(qc)
probs = sv.probabilities_dict()
print(f'Probabilities: {probs}')
plot_bloch_multivector(sv)
```
