```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from qiskit.primitives import StatevectorSampler 
from qiskit_aer import AerSimulator
import numpy as np

# Initialize the StatevectorSampler primitive for accurate local simulation
# This object is optimized for generating counts from a noiseless statevector.
sampler = StatevectorSampler()

def create_bomb_tester_circuit(bomb_status):
    """
    Creates the Mach-Zehnder interferometer circuit with the bomb in the upper path.
    """
    # Define circuit with one qubit and one classical bit (for result 'c')
    qc = QuantumCircuit(1, 1, name=f"Bomb_Tester_{bomb_status}")

    # 1. First 50/50 Beam Splitter
    qc.h(0)
    
    # --- The "Bomb" Component (Applied to the |1> component) ---
    if bomb_status == "live":
        # Live bomb: Z gate (180-degree phase shift)
        qc.z(0)
    elif bomb_status == "dud":
        # Dud bomb: Identity gate (qc.id)
        qc.id(0)
    else:
        raise ValueError("bomb_status must be 'live' or 'dud'.")
    
    # 2. Second 50/50 Beam Splitter
    qc.h(0)

    # 3. Measurement (to the default classical register 'c')
    qc.measure(0, 0)
    
    return qc

def run_experiment(bomb_status, shots=1000):
    """Runs the circuit using the StatevectorSampler and prints the results."""
    qc = create_bomb_tester_circuit(bomb_status)
    
    # Run the circuit using the Sampler primitive
    job = sampler.run([qc], shots=shots)
    result = job.result()
    
    # FIX: Access the counts. 
    # result[0].data is a DataBin. The counts for the single classical register 
    # (named 'c' by default) is accessed as an attribute, and the data is a BitArray 
    # which has the .get_counts() method.
    counts = result[0].data.c.get_counts()
    
    # Calculate probabilities
    p_0 = counts.get('0', 0) / shots * 100
    p_1 = counts.get('1', 0) / shots * 100
    
    print(f"--- Experiment with a {bomb_status.upper()} BOMB ---")
    print(qc.draw(output='text', fold=-1))
    print(f"\nMeasurement Counts (Shots: {shots}): {counts}")
    print(f"Probability of measuring |0⟩ (Safe/Success): {p_0:.1f}%")
    print(f"Probability of measuring |1⟩ (Explosion/Failure): {p_1:.1f}%")

    # Interpretation
    if bomb_status == "dud":
        # Theory: 100% |0>
        print("OUTCOME: ✅ **Dud Detected (SAFE)**. All photons recombined on the $|0\rangle$ path.")
    elif bomb_status == "live":
        # Theory: 50% |0> (Success), 50% |1> (Explosion)
        if p_0 > 0:
            print(f"OUTCOME: 🌟 **Success (INTERACTION-FREE)**. Measurement '|0⟩' occurred. This proves the bomb is **LIVE** without triggering it! (Probability ~{p_0:.1f}%)")
        if p_1 > 0:
            print("OUTCOME: 💥 **Explosion/Failure** also occurred on this run. Measurement '|1⟩' means the photon took the upper path and exploded the bomb.")


# --- Run the Scenarios ---
print("\n"+"#"*70)
print("### SCENARIO 1: The BOMB IS A DUD ###")
run_experiment("dud", shots=1000)

print("\n"+"#"*70)
print("### SCENARIO 2: The BOMB IS LIVE ###")
run_experiment("live", shots=1000)
print("\n"+"#"*70)
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import numpy as np

# Use the AerSimulator directly for the measurement process
simulator = AerSimulator(method='statevector')

def run_ev_experiment_correct(bomb_status, shots=1000):
    """
    Runs the corrected Elitzur-Vaidman experiment using conditional logic 
    to model the probabilistic explosion/collapse.
    """
    if bomb_status == "dud":
        # 1. DUD SCENARIO: The standard Mach-Zehnder interference
        qc = QuantumCircuit(1, 1, name="Dud_Test")
        qc.h(0)
        qc.id(0) # Dud (Identity)
        qc.h(0)
        qc.measure(0, 0)
        
        # Expected ideal result: 100% |0>
        
    elif bomb_status == "live":
        # 2. LIVE BOMB SCENARIO: The core probabilistic test
        # We must measure the path *before* the second H gate to model the explosion/collapse.
        qc = QuantumCircuit(1, 1, name="Live_Test_Probabilistic")
        qc.h(0)
        
        # This measurement models the bomb's interaction (explosion/collapse)
        # If the photon is in |1> (Upper Path), the bomb explodes (50% chance).
        qc.measure(0, 0)
        
        # We only continue the interferometer if the photon was NOT in the bomb path (m=0)
        # However, modeling the *non-detection* requires a different simulator setup.
        # The best we can do with standard Qiskit is model the conditional interference:
        
        # If the measurement result is 0 (Lower Path), apply the second H gate
        # Otherwise (measurement result is 1, Upper Path), the bomb exploded, and the experiment ends (no measurement at C or D).
        # We cannot model the 'No photon detected' (explosion) cleanly as a bitstring outcome.
        
        # We will use the standard EV circuit but rely on classical post-processing 
        # based on the measurement *before* the second H gate to model the explosion.
        pass # We will use the probabilistic analysis instead of the single circuit

    else:
        raise ValueError("Invalid bomb status.")


    if bomb_status == "dud":
        # Run the standard circuit
        t_qc = transpile(qc, simulator)
        job = simulator.run(t_qc, shots=shots)
        counts = job.result().get_counts()
        
        p_0 = counts.get('0', 0) / shots * 100
        p_1 = counts.get('1', 0) / shots * 100
        
        print(f"--- Experiment with a {bomb_status.upper()} BOMB ---")
        print(qc.draw(output='text', fold=-1))
        print(f"\nMeasurement Counts (Shots: {shots}): {counts}")
        print(f"P(|0⟩, Detector C): {p_0:.1f}%")
        print(f"P(|1⟩, Detector D): {p_1:.1f}%")
        print("OUTCOME: ✅ **Dud Detected (SAFE)**. Interference works perfectly (100% on C).")
        
    elif bomb_status == "live":
        # Simulate the first H gate: |ψ⟩ = (|0> + |1>)/√2
        # This is where the path choice is made. We use the Sampler to draw the path.
        qc_path = QuantumCircuit(1, 1, name="Live_Bomb_Path_Choice")
        qc_path.h(0)
        qc_path.measure(0, 0)
        
        job = sampler.run([qc_path], shots=shots)
        result = job.result()
        counts_path = result[0].data.c.get_counts()
        
        # Post-process the path choice
        path_1_count = counts_path.get('1', 0)  # Bomb Path (|1>)
        path_0_count = counts_path.get('0', 0)  # Safe Path (|0>)

        # 1. Explosion/No Photon: Occurs when the photon takes the |1> path.
        explosion_count = path_1_count 
        
        # 2. Interaction-Free Branch: Occurs when the photon takes the |0> path (Safe).
        # This safe path then hits the second H gate and splits 50/50.
        
        # Simulate the final split of the safe path
        qc_split = QuantumCircuit(1, 1, name="Final_Split")
        qc_split.h(0) # Starts in |0>, H splits to (|0> + |1>)/√2
        qc_split.measure(0, 0)
        
        job_split = simulator.run(qc_split, shots=path_0_count) # Only run for the safe-path count
        counts_split = job_split.result().get_counts()
        
        detector_C_count = counts_split.get('0', 0)
        detector_D_count = counts_split.get('1', 0)

        # Final Probabilities
        P_explosion = explosion_count / shots * 100
        P_C = detector_C_count / shots * 100
        P_D = detector_D_count / shots * 100
        
        print(f"--- Experiment with a {bomb_status.upper()} BOMB (Simulated Collapse) ---")
        print(f"Total Shots: {shots}")
        print(f"Path Choice Counts: {counts_path}")
        print("-" * 35)
        print("FINAL RESULTS:")
        print(f"P(No Photon Detected / Explosion): {P_explosion:.1f}% (Expected ~50.0%)")
        print(f"P(Detector C / Interaction-Free Success): {P_C:.1f}% (Expected ~25.0%)")
        print(f"P(Detector D / Safe Path): {P_D:.1f}% (Expected ~25.0%)")
        print("OUTCOME: 🌟 **Success Rate**: The probability of interaction-free detection (Detector C) is approximately 25%.")


# --- Run the Corrected Scenarios ---
print("\n"+"#"*70)
print("### SCENARIO 1: The BOMB IS A DUD (Corrected Model) ###")
run_ev_experiment_correct("dud", shots=1000)

print("\n"+"#"*70)
print("### SCENARIO 2: The BOMB IS LIVE (Corrected Probabilistic Model) ###")
run_ev_experiment_correct("live", shots=1000)
print("\n"+"#"*70)
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit.circuit.library import RYGate, IGate, ZGate 
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
import numpy as np

simulator = AerSimulator(method='statevector')

def run_multi_pass_ev_test(bomb_status, N_cycles=20, shots=1000):
    """
    Implements the high-yield EV bomb tester using the Quantum Zeno sequence, 
    including a final rotation to cancel the accumulated rotation for the Dud state.
    """
    qc = QuantumCircuit(1, 1, name=f"EV_Amplified_{bomb_status}")
    
    # Critical Angle: theta = pi / (2 * N)
    theta = np.pi / (2 * N_cycles)
    
    # Define the rotation instruction.
    Ry_theta = RYGate(theta)

    if bomb_status == "live":
        bomb_gate = ZGate() 
    else: # dud
        bomb_gate = IGate()
    
    # --- ZENO AMPLIFICATION LOOP ---
    # Total Rotation of the Dud state over N cycles is RY(pi)
    for _ in range(N_cycles):
        qc.append(Ry_theta, [0]) 
        qc.append(bomb_gate, [0])
        qc.append(Ry_theta, [0]) 
        
    # --- FINAL CORRECTION ---
    # The total accumulated angle for the Dud state is pi. 
    # To return |0> to |0>, we must apply RY(-pi).
    qc.ry(-np.pi, 0)
    
    # Final measurement
    qc.measure(0, 0)
    
    # Run the circuit on the simulator
    t_qc = transpile(qc, simulator)
    job = simulator.run(t_qc, shots=shots)
    counts = job.result().get_counts()

    fig = qc.draw(output="mpl", style="iqp", fold=60)
    
    # Calculate probabilities
    p_0 = counts.get('0', 0) / shots * 100
    p_1 = counts.get('1', 0) / shots * 100
    
    print(f"--- Amplified EV Test ({bomb_status.upper()} Bomb, N={N_cycles}) ---")
    print(f"P(|0⟩, Dud/Safe Detection): {p_0:.2f}%")
    print(f"P(|1⟩, Live/Failure Detection): {p_1:.2f}%")
    
    if bomb_status == "live":
        # EXPECTED: P(|1>) to be dominant (~100%)
        if p_1 > p_0:
             print(f"OUTCOME: 🌟 **Live Bomb Identified**. Success yield P(|1>) ≈ {p_1:.2f}% (High-yield confirmation).")
        else:
             print(f"OUTCOME: ❌ **Error**. Expected Live Bomb signal P(|1>) to be dominant.")
    else:
        # EXPECTED: P(|0>) to be dominant (~100%)
        if p_0 > p_1:
             print(f"OUTCOME: ✅ **Dud Bomb Identified**. Success yield P(|0>) ≈ {p_0:.2f}% (High-yield confirmation).")
        else:
             print(f"OUTCOME: ❌ **Error**. Expected Dud Bomb signal P(|0>) to be dominant.")

    return fig
    

# --- Run the High-Yield Scenarios ---
N = 20 # Number of cycles

print("\n"+"#"*70)
print(f"### SCENARIO 3: AMPLIFIED DUD BOMB (N={N}) ###")
run_multi_pass_ev_test("dud", N_cycles=N)

print("\n"+"#"*70)
print(f"### SCENARIO 4: AMPLIFIED LIVE BOMB (N={N}) ###")
run_multi_pass_ev_test("live", N_cycles=N)
print("\n"+"#"*70)

```

```python
from qiskit import QuantumCircuit, transpile
from qiskit.circuit.library import RYGate, IGate, ZGate
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display # For guaranteed plot display in Jupyter

# ----------------------------------------------------------------------
## ⚙️ CONFIGURATION AND SIMULATOR
# ----------------------------------------------------------------------

simulator = AerSimulator(method='statevector')
N_CYCLES = 20
SHOTS = 1000

# ----------------------------------------------------------------------
## 🧪 QUANTUM PROBE HELPER FUNCTION
# This is a streamlined version of your EV tester, returning only P(|1>).
# ----------------------------------------------------------------------

def run_quantum_probe(bomb_status, N_cycles=N_CYCLES, shots=SHOTS):
    """
    Runs the high-yield EV bomb tester and returns the probability of the 
    'Live Bomb' outcome (P(|1>)).
    """
    qc = QuantumCircuit(1, 1, name=f"Quantum_Probe")
    
    # Critical Angle
    theta = np.pi / (2 * N_cycles)
    Ry_theta = RYGate(theta)

    # Set the interaction gate based on status
    if bomb_status == "live":
        bomb_gate = ZGate() # Live Bomb: Applies phase shift
    else:
        bomb_gate = IGate() # Dud Bomb: No phase shift
    
    # ZENO AMPLIFICATION LOOP
    for _ in range(N_cycles):
        qc.append(Ry_theta, [0])
        qc.append(bomb_gate, [0])
        qc.append(Ry_theta, [0])
        
    # FINAL CORRECTION and Measurement
    qc.ry(-np.pi, 0)
    qc.measure(0, 0)
    
    # Run the circuit
    t_qc = transpile(qc, simulator)
    job = simulator.run(t_qc, shots=shots)
    counts = job.result().get_counts()
    
    # Return P(|1>) (Live Bomb / Hit probability)
    p_1 = counts.get('1', 0) / shots * 100
    return p_1

# ----------------------------------------------------------------------
## 🗺️ BATTLESHIP GAME SETUP
# The 10x10 grid (A-J, 1-10) is represented by a dictionary.
# ----------------------------------------------------------------------

# Helper function to convert (row_char, col_int) to 'A1' string
def to_coord(r, c):
    return f"{chr(65 + r)}{c + 1}"

# Define the 10x10 grid and place 5 ships.
# 'W' = Water, 'S' = Ship Segment
GRID_SIZE = 10
BATTLE_GRID = {}
for r in range(GRID_SIZE):
    for c in range(GRID_SIZE):
        BATTLE_GRID[to_coord(r, c)] = 'W' # Initialize to Water

# Manually place five ships (example placements)
SHIPS = {
    # Ship 1: Carrier (5 segments)
    'A1': 'S', 'A2': 'S', 'A3': 'S', 'A4': 'S', 'A5': 'S',
    # Ship 2: Battleship (4 segments)
    'C7': 'S', 'D7': 'S', 'E7': 'S', 'F7': 'S',
    # Ship 3: Cruiser (3 segments)
    'I2': 'S', 'I3': 'S', 'I4': 'S',
    # Ship 4: Submarine (3 segments)
    'E1': 'S', 'E2': 'S', 'E3': 'S',
    # Ship 5: Destroyer (2 segments)
    'J9': 'S', 'J10': 'S'
}

# Update the main grid with ship locations
BATTLE_GRID.update(SHIPS)

# ----------------------------------------------------------------------
## 🎯 BATTLESHIP QUANTUM PROBING LOGIC
# ----------------------------------------------------------------------

def probe_battleship_coordinate(grid, coord):
    """
    Probes a single coordinate using the EV Quantum Probe logic.
    Returns the probability of a 'Hit' (P(|1>)).
    """
    
    # 1. Determine the 'Bomb Status' based on the grid
    if grid.get(coord, 'W') == "S":
        status = "live" # Ship is present -> Live Bomb (Hit)
    else:
        status = "dud" # Water is present -> Dud Bomb (Miss)

    # 2. Run the quantum test and get the probability of a Live outcome (Hit)
    # The quantum code guarantees p_live is near 100% for 'live' and near 0% for 'dud'
    p_live = run_quantum_probe(status) 

    # The CLUE:
    print(f"Probe at {coord}: Probability of LIVE BOMB outcome (P|1>) is {p_live:.2f}%")
    return p_live

def run_full_grid_scan(grid):
    """
    Runs the quantum probe on every coordinate of the 10x10 grid and prints the results.
    """
    print("="*70)
    print("### ⚛️ BATTLESHIP FULL GRID QUANTUM SCAN RESULTS ###")
    print("="*70)
    
    # Print the header (A B C D E F G H I J)
    header = "   " + " ".join([chr(65 + i) for i in range(GRID_SIZE)])
    print(header)
    print("  " + "---"*GRID_SIZE)

    # Dictionary to store and print the results
    results_map = {}
    
    # Loop through all coordinates (Rows 1-10, Cols A-J)
    for r in range(GRID_SIZE):
        row_output = f"{r + 1:2d}|"
        for c in range(GRID_SIZE):
            coord = to_coord(r, c)
            
            # Run the probe and get the result (p_live)
            p_live = probe_battleship_coordinate(grid, coord)

            # Store the result symbol
            if p_live > 90: # Near 100% means a HIT
                symbol = "💥"
            else: # Near 0% means a MISS
                symbol = "🌊"
            
            results_map[coord] = symbol
            row_output += f" {symbol}"
        
        # Print the row results
        print(row_output)
    
    print("\nLegend: 💥 = HIGH P(|1>) (HIT) | 🌊 = LOW P(|1>) (MISS)")


# ----------------------------------------------------------------------
## 🚀 EXECUTION
# We need to run your original function to display the circuit one last time
# and then run the new Battleship game logic.
# ----------------------------------------------------------------------

# Running the original function to display the circuit (and its probabilities)
print("\n"+"#"*70)
print("### SCENARIO 4: AMPLIFIED LIVE BOMB CIRCUIT & PROBABILITY ###")
# We capture the figure but don't explicitly display it here to avoid repetition.
run_multi_pass_ev_test("live", N_cycles=N_CYCLES) 
# Note: You can call display(fig) manually if you want the circuit image.

# Running the new Battleship logic
run_full_grid_scan(BATTLE_GRID)

print("\n"+"#"*70)
```

```python
import random
from qiskit import qpy
from qiskit import QuantumCircuit # We need a placeholder circuit for the QPY format

# Define ship lengths
SHIP_SPECS = {
    'Carrier': 5,
    'Battleship': 4,
    'Cruiser': 3,
    'Submarine': 3,
    'Destroyer': 2
}
GRID_SIZE = 10
FILE_PATH = "battleship_oracle.qpy"

# Helper to convert coordinates
def to_coord(r, c):
    return f"{chr(65 + r)}{c + 1}"

def generate_empty_grid():
    grid = {}
    for r in range(GRID_SIZE):
        for c in range(GRID_SIZE):
            grid[to_coord(r, c)] = 'W' # 'W' for Water
    return grid

def place_ships_randomly():
    grid = generate_empty_grid()
    
    # 1. Randomly place ships
    for ship_name, length in SHIP_SPECS.items():
        placed = False
        while not placed:
            # Random starting point
            is_horizontal = random.choice([True, False])
            
            if is_horizontal:
                max_r = GRID_SIZE - 1
                max_c = GRID_SIZE - length
                r = random.randint(0, max_r)
                c = random.randint(0, max_c)
            else:
                max_r = GRID_SIZE - length
                max_c = GRID_SIZE - 1
                r = random.randint(0, max_r)
                c = random.randint(0, max_c)

            # Check for collisions and boundaries
            coords_to_place = []
            can_place = True
            for i in range(length):
                if is_horizontal:
                    coord = to_coord(r, c + i)
                else:
                    coord = to_coord(r + i, c)
                
                # Check if coordinate is already occupied
                if grid.get(coord, 'W') == 'S':
                    can_place = False
                    break
                
                coords_to_place.append(coord)
            
            # 2. Place the ship if collision-free
            if can_place:
                for coord in coords_to_place:
                    grid[coord] = 'S' # 'S' for Ship
                placed = True
    
    # 3. Create a Qiskit circuit to act as a container/oracle for the grid data
    # QPY requires a QuantumCircuit object, so we embed the dictionary inside its metadata.
    oracle_circuit = QuantumCircuit(1, name="BattleshipOracle")
    oracle_circuit.metadata = {'BATTLE_GRID_MAP': grid}
    
    return oracle_circuit, grid

# ----------------------------------------------------------------------
## 💾 EXECUTION BLOCK 1: GENERATE AND SAVE ORACLE
# ----------------------------------------------------------------------
print("### BLOCK 1: GENERATING RANDOM SHIPS AND SAVING ORACLE ###")
oracle_circuit, secret_grid = place_ships_randomly()

try:
    with open(FILE_PATH, "wb") as f:
        qpy.dump(oracle_circuit, f)
    print(f"✅ Ship placements successfully saved to '{FILE_PATH}' metadata.")
except Exception as e:
    print(f"❌ Error saving file: {e}")

# The secret grid is hidden from Block 2, but we print a preview here for debugging:
print("\n[DEBUG] SECRET SHIP LOCATIONS (Hiding this in Block 2):")
temp_output = "   " + " ".join([chr(65 + i) for i in range(GRID_SIZE)]) + "\n"
for r in range(GRID_SIZE):
    row_output = f"{r + 1:2d}|"
    for c in range(GRID_SIZE):
        coord = to_coord(r, c)
        row_output += f" {secret_grid[coord]}"
    temp_output += row_output + "\n"
print(temp_output)
```

```python
from qiskit import QuantumCircuit, transpile, qpy
from qiskit.circuit.library import RYGate, IGate, ZGate
from qiskit_aer import AerSimulator
import numpy as np
import os # For checking if the file exists

# ----------------------------------------------------------------------
## ⚙️ CONFIGURATION (Must match Block 1)
# ----------------------------------------------------------------------
simulator = AerSimulator(method='statevector')
N_CYCLES = 20
SHOTS = 1000
GRID_SIZE = 10
FILE_PATH = "battleship_oracle.qpy"
HIT_THRESHOLD = 90 # P(|1>) > 90% is considered a HIT

# Helper to convert coordinates
def to_coord(r, c):
    return f"{chr(65 + r)}{c + 1}"

# ----------------------------------------------------------------------
## 🧪 QUANTUM PROBE HELPER FUNCTION (Unchanged)
# ----------------------------------------------------------------------
def run_quantum_probe(bomb_status, N_cycles=N_CYCLES, shots=SHOTS):
    qc = QuantumCircuit(1, 1, name=f"Quantum_Probe")
    theta = np.pi / (2 * N_cycles)
    Ry_theta = RYGate(theta)
    bomb_gate = ZGate() if bomb_status == "live" else IGate()
    
    for _ in range(N_cycles):
        qc.append(Ry_theta, [0])
        qc.append(bomb_gate, [0])
        qc.append(Ry_theta, [0])
    
    qc.ry(-np.pi, 0)
    qc.measure(0, 0)
    
    t_qc = transpile(qc, simulator)
    job = simulator.run(t_qc, shots=shots)
    counts = job.result().get_counts()
    
    p_1 = counts.get('1', 0) / shots * 100
    return p_1

# ----------------------------------------------------------------------
## 🎯 BATTLESHIP PROBING AND VERIFICATION LOGIC
# ----------------------------------------------------------------------

def run_quantum_detection_and_verify():
    
    # 1. Load the true ship locations (Oracle)
    if not os.path.exists(FILE_PATH):
        print(f"❌ ERROR: Oracle file '{FILE_PATH}' not found. Run Block 1 first.")
        return
        
    try:
        with open(FILE_PATH, "rb") as f:
            loaded_oracle = qpy.load(f)[0] # QPY returns a list, take the first circuit
        true_grid = loaded_oracle.metadata['BATTLE_GRID_MAP']
        print(f"✅ Loaded secret grid from '{FILE_PATH}'. Starting quantum scan...")
    except Exception as e:
        print(f"❌ Error loading or decoding QPY file: {e}")
        return

    # 2. Run Quantum Detection
    quantum_guesses = {}
    correct_guesses = 0
    total_segments = 0

    print("\n" + "="*70)
    print("### ⚛️ BLOCK 2: QUANTUM DETECTION AND GUESSES ###")
    print("="*70)
    
    # Header for the detection map
    header = "   " + " ".join([chr(65 + i) for i in range(GRID_SIZE)])
    print(header)
    print("  " + "---"*GRID_SIZE)

    # Loop through all coordinates
    for r in range(GRID_SIZE):
        row_output = f"{r + 1:2d}|"
        for c in range(GRID_SIZE):
            coord = to_coord(r, c)
            
            # Determine the status for the probe based on the truth (emulating a physical interaction)
            status = "live" if true_grid[coord] == "S" else "dud"

            # RUN THE QUANTUM PROBE
            p_live = run_quantum_probe(status) 

            # 3. Determine the Quantum Guess
            if p_live > HIT_THRESHOLD:
                guess = "S" # Quantum system guesses Ship
                symbol = "💥"
            else:
                guess = "W" # Quantum system guesses Water
                symbol = "🌊"
            
            quantum_guesses[coord] = guess
            row_output += f" {symbol}"
            
            # 4. Verification Check
            if guess == true_grid[coord]:
                correct_guesses += 1
            if true_grid[coord] == "S":
                total_segments += 1
                
        print(row_output)
    
    # 5. Final Verification Results
    total_coords = GRID_SIZE * GRID_SIZE
    accuracy = (correct_guesses / total_coords) * 100
    
    print("\n" + "="*70)
    print("### 📊 VERIFICATION RESULTS ###")
    print(f"Total Coordinates Probed: {total_coords}")
    print(f"Total Ship Segments (Truth): {total_segments}")
    print(f"Correct Quantum Guesses: {correct_guesses} / {total_coords}")
    print(f"Overall Accuracy: {accuracy:.2f}% (Should be near 100%)")
    print("="*70)

# ----------------------------------------------------------------------
## 🚀 EXECUTION BLOCK 2: RUN DETECTION AND VERIFY
# ----------------------------------------------------------------------
run_quantum_detection_and_verify()
```

```python
import random
from qiskit import QuantumCircuit, qpy
from qiskit.circuit.library import ZGate

# ----------------------------------------------------------------------
## ⚙️ CONFIGURATION & UTILITIES
# ----------------------------------------------------------------------
GRID_SIZE_2x2 = 2 # 2x2 grid requires 2 address qubits
FILE_PATH = "battleship_grover_oracle.qpy"

# Coordinates: 'A1', 'A2', 'B1', 'B2'
COORDS = [f"{chr(65 + r)}{c + 1}" for r in range(GRID_SIZE_2x2) for c in range(GRID_SIZE_2x2)]
ADDRESS_QUBITS = 2 # N=2 Address Qubits for 4 states (00, 01, 10, 11)

def random_place_single_ship():
    """Places a single ship segment randomly in the 2x2 grid."""
    
    # 1. Randomly select the 'Hit' coordinate
    hit_coord = random.choice(COORDS) 
    
    # 2. Map this coordinate to its address state (e.g., 'B2' -> 11)
    # The states are |00> (A1), |01> (A2), |10> (B1), |11> (B2)
    address_map = {coord: bin(i)[2:].zfill(ADDRESS_QUBITS) for i, coord in enumerate(COORDS)}
    hit_state = address_map[hit_coord] # e.g., '11'
    
    # 3. Create the Black Box Oracle (Z-gate controlled by the hit state)
    # The oracle marks the 'hit' state by applying a phase shift (-1).
    # We use a multi-controlled Z gate (MCMT) for this.
    # We need 2 address qubits and 1 ancilla qubit for the phase flip.
    
    # The oracle is a black box that flips the phase of the hit state.
    # For a hit_state='11', the circuit is a simple CZ on the two address qubits (Z-gate on control).
    # For a hit_state='10', we use X gates to invert the target qubit (qubit 1) before and after the CZ.
    
    # Since Qiskit's MCMTGate doesn't directly support phase kickback easily for arbitrary controls,
    # we'll use a simpler phase oracle that marks the '11' state for simplicity:
    oracle = QuantumCircuit(ADDRESS_QUBITS, name="BattleshipOracle")
    
    if hit_state == '00':
        # Mark |00> state (Apply X on both controls)
        oracle.x([0, 1])
        oracle.cz(0, 1)
        oracle.x([0, 1])
    elif hit_state == '01':
        # Mark |01> state (Apply X on control 0)
        oracle.x(0)
        oracle.cz(0, 1)
        oracle.x(0)
    elif hit_state == '10':
        # Mark |10> state (Apply X on control 1)
        oracle.cz(0, 1)
        # No X on control 0 needed if using CZ on 1st qubit (the qiskit way)
        oracle.x(1) # Apply X on control 1 (the one that should be 0)
        oracle.cz(0, 1)
        oracle.x(1)

    elif hit_state == '11':
        # Mark |11> state
        oracle.cz(0, 1)
        
    return oracle, hit_coord

# ----------------------------------------------------------------------
## 💾 EXECUTION BLOCK 1: CREATE AND SAVE ORACLE
# ----------------------------------------------------------------------
print("### BLOCK 1: CREATING BLIND ORACLE AND SAVING QPY ###")

# Create the black-box oracle and get the secret answer
oracle_circuit, secret_hit_coord = random_place_single_ship()

# Save the oracle (the black box) to a QPY file
try:
    with open(FILE_PATH, "wb") as f:
        # We save the circuit itself, which is the black box for Block 2
        qpy.dump([oracle_circuit], f) 
    print(f"✅ Oracle (Black Box) saved to '{FILE_PATH}'.")
    print(f"[DEBUG] SECRET HIT LOCATION (Hiding in Block 2): {secret_hit_coord}")
except Exception as e:
    print(f"❌ Error saving QPY file: {e}")

# Note: The true location is not in the QPY metadata—only the circuit is saved.
```

```python
import random
from qiskit import QuantumCircuit, qpy
from qiskit.circuit.library import ZGate, MCMTGate
import numpy as np
import os

# ----------------------------------------------------------------------
## ⚙️ CONFIGURATION & UTILITIES (4x4 Grid - NO ANCILLA)
# ----------------------------------------------------------------------
ADDRESS_QUBITS = 4 
FILE_PATH = "battleship_grover_oracle_4x4_stable.qpy"

# Coordinates: 'A1' through 'D4' (16 states total)
COORDS = [f"{chr(65 + r)}{c + 1}" for r in range(4) for c in range(4)]

def random_place_single_ship():
    """Places a single ship segment randomly and creates the 4-qubit oracle circuit."""
    
    # 1. Randomly select the 'Hit' coordinate
    hit_coord = random.choice(COORDS) 
    address_map = {coord: bin(i)[2:].zfill(ADDRESS_QUBITS) for i, coord in enumerate(COORDS)}
    hit_state = address_map[hit_coord] # e.g., '0011'
    
    # 2. Create the Black Box Oracle (acting on 4 qubits)
    oracle = QuantumCircuit(ADDRESS_QUBITS, name="BattleshipOracle")
    
    # Identify which qubits should be inverted (i.e., should be |0> in the target state)
    # The list is indexed LSB first (Q0, Q1, Q2, Q3)
    controls_to_invert = [i for i, bit in enumerate(hit_state[::-1]) if bit == '0']
    
    # 3a. Pre-inversion (on all qubits 0-3)
    if controls_to_invert:
        oracle.x(controls_to_invert)
    
    # 3b. Apply the 4-Controlled Z (CCCCZ) using MCMTGate
    # We need 4 controls for 16 states. MCMTGate is parameterized as (Target Gate, Num Controls, Num Targets)
    # Since we need 4 total qubits, we use 3 controls and 1 target (Q0, Q1, Q2 controls; Q3 target).
    # This acts as a 4-qubit controlled-Z on the whole address register.
    
    mcmt_z_gate = MCMTGate(ZGate(), num_ctrl_qubits=ADDRESS_QUBITS - 1, num_target_qubits=1)
    
    # Append to all 4 qubits (Q0-Q3). Q0, Q1, Q2 are controls, Q3 is the target of the Z gate.
    oracle.append(mcmt_z_gate, list(range(ADDRESS_QUBITS)))
    
    # 3c. Post-inversion
    if controls_to_invert:
        oracle.x(controls_to_invert)
        
    return oracle, hit_coord, hit_state

# ----------------------------------------------------------------------
## 💾 EXECUTION BLOCK 1: CREATE AND SAVE ORACLE
# ----------------------------------------------------------------------
print("### BLOCK 1: GENERATING STABLE ORACLE (4x4, 4-Qubit) AND SAVING QPY ###")

oracle_circuit, secret_hit_coord, secret_state_binary = random_place_single_ship()

try:
    with open(FILE_PATH, "wb") as f:
        qpy.dump([oracle_circuit], f) 
    print(f"✅ Oracle (Black Box) saved to '{FILE_PATH}'.")
    print(f"[DEBUG] SECRET HIT LOCATION: {secret_hit_coord} (Binary: {secret_state_binary})")
except Exception as e:
    print(f"❌ Error saving QPY file: {e}")
```

```python
from qiskit import QuantumCircuit, transpile, qpy
from qiskit_aer import AerSimulator
from qiskit.circuit.library import MCMTGate, ZGate
import numpy as np
import os

# ----------------------------------------------------------------------
## ⚙️ CONFIGURATION (Must match Block 1)
# ----------------------------------------------------------------------
simulator = AerSimulator(method='statevector')
ADDRESS_QUBITS = 4 
FILE_PATH = "battleship_grover_oracle_4x4_stable.qpy"
ITERATIONS = 3     # Optimal iterations for N=16 states (round(pi/4 * sqrt(16)) = 3)

# Coordinates map for verification
COORDS = [f"{chr(65 + r)}{c + 1}" for r in range(4) for c in range(4)]
address_map_inv = {bin(i)[2:].zfill(ADDRESS_QUBITS): coord for i, coord in enumerate(COORDS)}

# ----------------------------------------------------------------------
## ⚛️ GROVER SEARCH COMPONENTS
# ----------------------------------------------------------------------

def build_diffuser(n_qubits):
    """
    Builds the Grover Diffuser (Amplifier) on n_qubits using the standard H X MCZ X H structure.
    """
    qc = QuantumCircuit(n_qubits, name="Diffuser")
    
    # 1. H and X
    qc.h(range(n_qubits))
    qc.x(range(n_qubits))
    
    # 2. Multi-Controlled Z (MCZ)
    mcmt_z_gate = MCMTGate(ZGate(), num_ctrl_qubits=n_qubits - 1, num_target_qubits=1)
    qc.append(mcmt_z_gate, list(range(n_qubits)))
    
    # 3. X and H
    qc.x(range(n_qubits))
    qc.h(range(n_qubits))
    return qc

def run_ev_search():
    
    # 1. Load the Black Box Oracle
    if not os.path.exists(FILE_PATH):
        print(f"❌ ERROR: Oracle file '{FILE_PATH}' not found. Run Block 1 first.")
        return None, None
    
    try:
        with open(FILE_PATH, "rb") as f:
            oracle_circuit = qpy.load(f)[0]
    except Exception as e:
        print(f"❌ Error loading QPY file: {e}")
        return None, None

    # 2. Setup the Search Circuit
    qc = QuantumCircuit(ADDRESS_QUBITS, ADDRESS_QUBITS, name="Grover_Search")
    
    # Initialize the superposition state
    qc.h(range(ADDRESS_QUBITS))
    
    # Build the Diffuser
    diffuser = build_diffuser(ADDRESS_QUBITS)
    
    # 3. Apply the Grover Iteration
    for _ in range(ITERATIONS):
        qc.append(oracle_circuit.to_instruction(), range(ADDRESS_QUBITS))
        qc.append(diffuser.to_instruction(), range(ADDRESS_QUBITS))

    # 4. Final Measurement
    qc.measure(range(ADDRESS_QUBITS), range(ADDRESS_QUBITS))
    
    # --- Run ---
    qc_decomposed = qc.decompose()
    
    print("\n✅ Circuit composed and running simulation...")

    # 5. Run the Simulation
    t_qc = transpile(qc_decomposed, simulator)
    job = simulator.run(t_qc, shots=1024)
    counts = job.result().get_counts()
    
    # 6. Determine the Guess
    most_likely_state = max(counts, key=counts.get)
    quantum_guess = address_map_inv[most_likely_state]
    
    return quantum_guess, counts

# ----------------------------------------------------------------------
## 🚀 EXECUTION BLOCK 2: RUN DETECTION AND VERIFY
# ----------------------------------------------------------------------

print("\n### BLOCK 2: RUNNING QUANTUM DETECTION (BLIND SEARCH) ###")
quantum_guess_coord, results = run_ev_search()

if quantum_guess_coord and results:
    prob_guess = results.get(max(results, key=results.get), 0) / 1024 * 100

    print("\n" + "="*70)
    print("### 🎯 QUANTUM DETECTION RESULT ###")
    print(f"Quantum Algorithm's Guess (Most Likely State): **{quantum_guess_coord}**")
    print(f"Probability of Guess: **{prob_guess:.2f}%**")

    print("\n### 🔎 VERIFICATION AGAINST SECRET ORACLE ###")
    print("Check this guess against the SECRET HIT LOCATION printed in BLOCK 1.")
    print("======================================================================")
```
